#### Metoda SIMPLEX

Celem jest minimalizacja kosztw przy zachowaniu minimalnej ale koniecznej iloci rodkw odywczych


$ \bbox[5px,border:1px solid red]
{
F(x_1,x_2,x_3) = 100x_{1} + 50x_{2}+ 200x_{3} \to  min 
\qquad (funkcja-celu)  
} $

#### $
\left\{
\begin{align} 
{28x_1+14x_2+10x_3}   & \leqslant 14 & (1)\\ 
{10x_1+12x_2+6x_3}  & \leqslant 8 & (1)\\  
{30x_1+20x_2+30x_3} & \geqslant 25 & (2)\\
{10x_1+10x_2+15x_3} & \geqslant 12 & (2)\\
{x_1+x_2+x_3}& =1  & (2)\\
{x_1} & \geqslant 0 & (3)\\
{x_2} & \geqslant 0 & (4)\\
{x_3} & \geqslant 0 & (4)\\
\end{align}
\right. 
$

Gdy jest znak >= zmieniamy znak liczb przy x

Gdy funkcja celu dy do max to zmienamy znaki liczb przy x

$$
G = \begin{bmatrix}
28   &  14  &  10  \\
10   &  12  &  6   \\
-30  & -20  & -30  \\
-10  & -10  & -15  \\
 1   &  1   &  1   \\
 -1  &  0   &  0   \\
  0  &  -1  &  0   \\ 
  0  &  0  &  -1   \\ 
\end{bmatrix}
$$


$$
h = \begin{bmatrix}
14  \\
8  \\
-25   \\
-12   \\
1 \\
0 \\
0 \\
0 
\end{bmatrix}
$$

$$
c = \begin{bmatrix}
100 \\
50 \\
200
\end{bmatrix}
$$

```python
from cvxopt import matrix
import numpy as np

c = matrix([[100.0,50.0,200.0]]) ## <== przy Maxymalizacji oba znaki maj by ujemne
G = matrix([[28.0,10.0,-30.0,-10.0,1.0,-1.0,0.0,0.0],[14.0,6.0,-20.0,-10.0,1.0,0.0,-1.0,0.0],[10.0,6.0,-30.0,-15.0,1.0,0.0,0.0,-1.0]]) 
h = matrix([14.0,8.0,-25.0,-12.0,1.0,0.0,0.0,0.0])

from cvxopt import solvers

solvers.options['show_progress'] = False  ## nie chce oglda wydruku z oblicze
sol = solvers.lp(c, G, h)
zysk = sol['primal objective']


print('Zysk dla ustawie optymalnych: %.0f z' % zysk)
x =np.array(sol['x']).round(1)
print()
print('Macierz wielkoci optymalnych: x1, x2, x3, x4')
print()
print(x)
```
