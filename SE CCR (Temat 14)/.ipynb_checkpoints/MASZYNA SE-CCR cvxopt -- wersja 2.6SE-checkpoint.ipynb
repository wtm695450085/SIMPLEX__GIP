<span style="font-size: 200%;color:white;background:#1155cc"> 2 </span>
<span style="font-size: 200%;color:white;background:#f4cccc">  Liczymy pierwszy bank - tworzymy wektor nakadw
    
    pierwsze cztery wiersze oznaczaj nakady, ostatnie dwa wiersze oznaczaj efekty 

```python
BB = 'B7' # <-- TU WPISUJEMY NAZW BANKU, KTRY BDZIE ANALIZOWANY 
RR = 4   # <-- ILE JEST W BAZIE WIERSZY Z NAKADAMI
df  # nazwa bazy danych
```

```python
def matrix(name,BB,RR,df):
    
    import numpy as np
    from cvxopt import matrix
    from cvxopt import solvers
    
    global x, wynik
    global dfv
    
## ---WEK---------------tworzymy wektor nakadw dla wybranego obiektu---
    WEK = df[BB][:RR]
    WEK = np.array(WEK)  
    WEK = WEK*-1
    mop = np.array([0,0])  
    WEK = np.append(WEK, mop, axis=0) 
## ---KOR---------------tworzymy wektor efektw dla wybranego obiektu----
    KOR = df[BB][RR:]
    KOR = np.array(KOR)  
    KOR = KOR*-1
    ze = np.zeros(RR)                    # <--- UWAGA! Liczba wierszy z kosztami
    KOR = np.append(ze,KOR, axis=0)    
## ---NAKADY--Tworzymy macierz nakadw dla wszystkich--------------
    NAKADY = df[:RR] 
    NAKADY = np.array(NAKADY)  
    NAKADY = NAKADY.T
    NAKADY[column_values.index(BB)]=0    ### FUNKCJA SPECJALNA DLA SE super-efficiency
##---EFEKTY--Tworzymy wektor efektw dla wszystkich-------------------
    EFEKTY = df[RR:]
    EFEKTY = np.array(EFEKTY)
    EFEKTY = EFEKTY.T*-1
    EFEKTY[column_values.index(BB)] = 0   ### FUNKCJA SPECJALNA DLA SE super-efficiency
##---czymy MACIERZ efektw i nakadw dla wszystkich----------------
    GN = np.append(NAKADY,EFEKTY, axis=1)     
##--Dodajemy do macierzy efe. i nak., wektor specjalny dla wybranego obiektu -------
    w,k =df.shape
    WEK = WEK.reshape(1,w)             # <--- UWAGA! Liczba wierszy z kosztami + wynikami
    GN = np.append(WEK,GN, axis=0) 
    
##---Macierz warunkw znakowych ----------------------------------------
    w,k = GN.shape
    ZERO = np.eye(w-1, k=0, dtype=int) *-1
    w,k = ZERO.shape
    ze = np.zeros(w)
    ze = ze.reshape(1,w)
    ZERO = np.append(ze,ZERO, axis=0) 
    w,k = ZERO.shape
    zk = np.zeros(w)
    
    # Dodajemy wektor z jedynk k
    zk =zk.reshape(w,1)
    GF = np.append(ZERO,zk, axis=1) 
    w, k = GF.shape
    GF[w-1,k-1]=1
### ----OSTATECZNA MACIERZ ----czymy wszystkie macierze
    GK = np.append(GN,GF, axis=1) 
    G = GK.tolist()
### ----Tworz wektor celu c------------------------
    w,k = GK.shape
    C = np.zeros(w)
    C[0]=1
    C = C.tolist()
### ----Tworz wektor wynikw h--------------------
    a = KOR.shape
    nt = k-a[0]
    nv = np.zeros(nt)
    h = np.append(KOR,nv, axis=0) 
    h = h.tolist()
###----------------------------------    
  
    c = matrix(C) ## <== przy Maxymalizacji oba znaki maj by ujemne
    G = matrix(G)
    h = matrix(h)

    solvers.options['show_progress'] = False  ## nie chce oglda wydruku z oblicze
    sol = solvers.lp(c, G, h)
    zysk = sol['primal objective']


    #print('Zysk dla ustawie optymalnych: %.0f z' %-zysk)
    x =np.array(sol['x']).round(3)
    x[column_values.index(BB)+1]=0    ## <-- wstawka kasujca wartoci po przejtnej tylko dla SE CCR
        
    #na = ['Bank Handlowy', 'BRE Bank', 'BIG Bank', 'Citibank', 'PKO BP', 'Bank Gospodarstwa Krajowego', 'Bank lski', 'PeKaO S.A.', 'Opel Bank', 'Bank Gospodarki ywnociowej']
    na = ['B1', 'B2', 'B3', 'B4', 'B5', 'B6', 'B7', 'B8', 'B9', 'B10']
    
    dfv = pd.DataFrame({'model':name, 'k':x[0], na[0]:x[1], na[1]:x[2], na[2]:x[3], na[3]:x[4], na[4]:x[5], na[5]:x[6], na[6]:x[7], na[7]:x[8], na[8]:x[9], na[9]:x[10]})
    
    return dfv
     
```

<span style="font-size: 200%;color:white;background:#1155cc"> 3</span>
<span style="font-size: 200%;color:white;background:#a2c4c9">    Podstawiam dane do funkcji
    
    UWAGA! Liczba wierszy z kosztami 

```python
def Cohen(BANKI,name,df):

    global SE_CCR
    
    SE_CCR = []
    
    for e, n in  zip(BANKI,name):        

        results = matrix(n,e,4,df)    # <--- UWAGA! Liczba wierszy z kosztami
        SE_CCR.append(results)
    SE_CCR = pd.concat(SE_CCR)   
          
    return SE_CCR
```
