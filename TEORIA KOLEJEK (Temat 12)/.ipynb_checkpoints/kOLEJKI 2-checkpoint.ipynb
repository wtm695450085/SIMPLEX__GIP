```python
def factorial(n): 
    return 1 if (n==1 or n==0) else n * factorial(n - 1);  
```

# 1. Prawdopodobiestwo, 偶e nie bdzie kolejki

<span style="font-size: 180%;color:black"> 
$${
{\displaystyle\sum_{i=0}^{r-1} \frac{^i}{i!}}
}\qquad $$


<span style="font-size: 180%;color:black"> 
$${
{\displaystyle\sum_{i=0}^{r-1} \frac{^i}{i!}}= \frac{0.9^{(1-1)}}{{(1-1)}!}+\frac{0.9^{(2-1)}}{{(2-1)}!}+\frac{0.9^{(3-1)}}{{(3-1)}!}=1+0.9+0.405 \approx 2.305
}\qquad $$

```python
def dzings(媳,r):
    
    a1 = (媳**0)/factorial(0) # gdy kana贸w jest r=1 wtedy NT = 1
    a2 = a1+(媳**1)/factorial(1) # gdy kana贸w jest r=2 wtedy NT = a0
    a3 = a2+(媳**2)/factorial(2) # gdy kana贸w jest r=3 wtedy NT = a0 + a1 
    a4 = a3+(媳**3)/factorial(3) # gdy kana贸w jest r=4 wtedy NT = a0 + a1 + a2 
    a5 = a4+(媳**4)/factorial(4) # gdy kana贸w jest r=5 wtedy NT = a0 + a1 + a2 + a3 
    a6 = a5+(媳**5)/factorial(5) # gdy kana贸w jest r=6 wtedy NT = a0 + a1 + a2 + a3 + a4
    a7 = a6+(媳**6)/factorial(6) # gdy kana贸w jest r=7 wtedy NT = a0 + a1 + a2 + a3 + a4 + a5
    a8 = a7+(媳**7)/factorial(7) # gdy kana贸w jest r=8 wtedy NT = a0 + a1 + a2 + a3 + a4 + a5 + a6
    a9 = a8+(媳**8)/factorial(8) # gdy kana贸w jest r=9 wtedy NT = a0 + a1 + a2 + a3 + a4 + a5 + a6 + a7
    a10 = a9+(媳**9)/factorial(9) # gdy kana贸w jest r=9 wtedy NT = a0+a1+a2+a3+a4+a5+a6+a7+a8


    if r == 1:
        NT = a1
    else:
        if r == 2:
            NT = a2
        else:
            if r == 3:
                NT = a3
            else:   
                if r == 4:
                    NT = a4
                else:   
                    if r == 5:
                        NT = a5
                    else:                       
                        if r == 6:
                            NT = a6
                        else:
                            if r == 7:
                                NT = a7
                            else:
                                if r == 8:
                                    NT = a8
                                else:
                                    if r == 9:
                                        NT = a9
                                    else:
                                        if r == 10:
                                            NT = a10
    return NT
```

<span style="font-size: 180%;color:black"> Prawdopodobiestwo, 偶e nie bdzie kolejki przed dokami

<span style="font-size: 180%;color:black">
$${\bbox[#f3f3f3,16px,border:1px solid black]
{p(n=0) =\frac{1}{\displaystyle\sum_{i=0}^{r-1} \frac{^i}{i!}+\frac{^r}{{\left(r-{}\right)}{\left(r-1\right)!}}}}
}\qquad (2)$$

```python
def prawd_braku_kolejki(媳,r):
    
    pbk=1/(NT+(((媳**r)/((r-媳)*factorial(r-1)))))
    
    return pbk
```

### Prawdopodobiestwo, 偶e bdzie kolejka przed dokami 

        n - ilo samochod贸w

<span style="font-size: 180%;color:black">
$$\\   
 \bbox[#f3f3f3,12px,border:1px solid black]
{p(n) =\begin{cases} 
\frac{{媳^n}{p(n=0)}}{n!},  & \text{if $n\le r$} \\[2ex] 
\frac{{r^{r-n}}{媳^n}{p(n=0)}}{r!},  & \text{if $n\gt r$} \\[2ex] 
\end{cases}
} \qquad (3)
$$

```python
def prawd_kolejki(媳,r,n,PBK):
    
        if n <= r:
            pt=(((媳**n)*PBK)/factorial(n))
        else: 
            pt=(((r**(r-n))*(媳**(n))*PBK))/(factorial(r))
        
        return pt      
```

## REDNIA ILI CI呕ARWEK czekajcych na podje藕dzie

<span style="font-size: 180%;color:black">

$$\ 
 \bbox[#f3f3f3,12px,border:1px solid black]
{
 =\frac{{^{r+1}{p(n=0)}}}{{(r-)}^2{(r-1)!}}
} \qquad (4)
$$

```python
def ile_sztuk_czeka(媳,r,PBK):

    ISC = ((**(r+1))*PBK)/(((r-)**2)*factorial(r-1))
    
    return ISC
```

## Prawdopodobiestwo, 偶e spdzimy t czasu w kolejce

<span style="font-size: 180%;color:black">
$$  
 \bbox[#f3f3f3,12px,border:1px solid black]
{
p(t>t_0) ={{p{(n>r-1)}}{e^{渭t_0(r-)}}}
} \qquad (5)
$$

```python
def prawdop_czasu(渭,r,,PBK, P1, t):
      
    import numpy as np
       
    PC = (1-(PBK+P1))*(np.e**(-渭*t*(r-)))
    
    return PC
```
