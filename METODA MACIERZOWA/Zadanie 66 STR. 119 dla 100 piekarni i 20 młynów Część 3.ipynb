```python
from pulp import *

km = makeDict([magazyny,piekarnie],KM,0)
print(km)
```

```python
prob = LpProblem("Zboża",LpMinimize)
print(prob)
```

```python
# Creates a list of tuples containing all the possible routes for transport
Routes = [(w,b) for w in magazyny for b in piekarnie]
print(Routes)
```

```python
# A dictionary called 'Vars' is created to contain the referenced variables(the routes)
vars = LpVariable.dicts("Route",(magazyny,piekarnie),0,None,LpInteger)
vars
```

```python
# The objective function is added to 'prob' first
prob += lpSum([vars[w][b]*km[w][b] for (w,b) in Routes]), "Sum_of_Transporting_Costs"
prob
```

```python
# The supply maximum constraints are added to prob for each supply node (warehouse)
for w in magazyny:
    prob += lpSum([vars[w][b] for b in piekarnie])<=supply[w], "Sum_of_Products_out_of_Warehouse_%s"%w
prob
```

```python
# The demand minimum constraints are added to prob for each demand node (bar)
for b in piekarnie:
    prob += lpSum([vars[w][b] for w in magazyny])>=demand[b], "Sum_of_Products_into_Bar%s"%b
prob
```

```python
# The problem data is written to an .lp file
prob.writeLP("zad66str119.lp")
```

```python
# The problem is solved using PuLP's choice of Solver
print(prob.solve())
print("Status:", LpStatus[prob.status])
```

```python
# Each of the variables is printed with it's resolved optimum value
GT =[]

for v in prob.variables():
    if v.varValue > 0:    ##<---Tylko te miejsca gdzie jest jakiś ruch 
        print(v.name, "-----------", v.varValue)
        GT.append(v)       #<-- Tworzę listę zmiennych dyskretnych 
   
```

```python
x = {}
for v in prob.variables():
    if v.varValue > 0:    ##<---Tylko te miejsca gdzie jest jakiś ruch 
        x[v.name] =  v.varValue
```

```python

PKP = pd.DataFrame.from_dict(x, orient="index").reset_index(0) 
PKP.columns = ['Nazwa_trasy', 'Ładunek w tonach'] 
PKP
```

### Dziele kolumny i dopisuje limity

```python
PKP['Route_'], PKP['Magazyn'], PKP['Piekarnia'] = PKP['Nazwa_trasy'].str.rsplit('_', 2).str
PKP
```

```python
PKP['Suply'] = PKP['Magazyn'].map(supply)
PKP['Demand'] = PKP['Piekarnia'].map(demand)
PKP
```
