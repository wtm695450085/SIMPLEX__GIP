```python
import pandas as pd

cc =["W_1", "W_2"]
aaa = [6, 6]
bbb = [10, 5]

df = pd.DataFrame({'Środki': cc, 'S_1': aaa, 'S_2': bbb  })
df
```

```python
import pandas as pd
from pulp import *

prob = LpProblem("METODA_PANDAS",LpMaximize)
```

### Tworzę listę podstawową

```python
wyroby = list(df['Środki'])
wyroby
```

### Tworzę listę zysku

### Tworzę słownik surowców

```python
S_1 = dict(zip(wyroby,df['S_1']))
S_2 = dict(zip(wyroby,df['S_2']))
```

```python
S_1
```

```python
S_2
```

### Określam dolną granicę

słownik zmiennych ilość wyrobów z dolną granicą = 0 i kategorią ciągłą, tj. Rozwiązanie optymalizacyjne może przyjąć dowolną wartość liczbową większą od zera.

```python
W_prod = LpVariable.dicts(name ="prod",indexs=wyroby,lowBound =0,cat='Continuous')
W_prod
```

```python
print(prob)
```

### Funkcja celu

```python
prob += lpSum([W_prod[i] for i in wyroby]), "Total Cost of the balanced diet"
```

```python
print(prob)
```

```python
import pandas as pd

cc =["W_1", "W_2", "W_3", "W_4", "W_5"]
aaa = [6,  6,4,7,4]
bbb = [10, 5,2,4,11]
ccc = [4,  2,3,5,1]
ddd = [9,  8,5,6,3]

df = pd.DataFrame({'Środki': cc, 'S_1': aaa, 'S_2': bbb,  'S_3': ccc,'S_4': ddd  })
df
```

```python
import pandas as pd
from pulp import *

prob = LpProblem("METODA_PANDAS",LpMaximize)
```

### Tworzę listę podstawową

```python
wyroby = list(df['Środki'])
wyroby
```

### Tworzę słownik surowców

```python
S_1 = dict(zip(wyroby,df['S_1']))
S_2 = dict(zip(wyroby,df['S_2']))
S_3 = dict(zip(wyroby,df['S_3']))
S_4 = dict(zip(wyroby,df['S_4']))
```

```python
S_1
```

```python
S_2
```

### Określam dolną granicę

słownik zmiennych ilość wyrobów z dolną granicą = 0 i kategorią ciągłą, tj. Rozwiązanie optymalizacyjne może przyjąć dowolną wartość liczbową większą od zera.

```python
W_prod = LpVariable.dicts(name ="prod",indexs=wyroby,lowBound =0,cat='Continuous')
W_prod
```

```python
print(prob)
```

### Funkcja celu

```python
prob += lpSum([W_prod[i] for i in wyroby]), "Total Cost of the balanced diet"
```

```python
print(prob)
```
